## Descriptive Statistics: 
Descriptive Statistics is nothing but by using Numerical methods and Visualization techniques, summarize the sample data. As we are summarizing the data in one group(sample), hence there is no uncertainty in data.
### Numerical Methods in Descriptive Statistics:
#### 1.	Measures of Central Tendency: 
Describes the central position of a frequency distribution for a group of data.
##### a)	Mean: 
Mean is average of the dataset. 
- Disadvantage of Mean is Outliers. If we have outliers, then go to Median or remove the outliers and calculate the Mean.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import sqrt
```

```python
data = pd.read_csv('Datasets/CarData.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
# Check the average price of selling price
np.mean(data['Selling_Price'])
```

##### b)	Median: 
Median is middle number after sorting the dataset. 
- If the number of records is Odd number, consider middle number as Median
- If the number of records is Even number, consider average of 2 middle numbers as Median


```python
np.median(data['Present_Price'])
```

##### c)	Mode: 
Mode is most occurrence values
- Bi model Mode: If 2 numbers of occurring high frequency as same number of times.
- Multi model Mode: If more than 2 numbers occurring high frequency as same number of times.

```python
x=data['Year'].mode()
x.iloc[0]
```

#### 2.	Measures of spread: 
How far out from the center, does the data extend? You can use the range or standard deviation to measure the dispersion. A low dispersion indicates that the data points are more tightly around the center. Higher dispersion signifies that data points fall further away from the center. 
##### a)	Range: 
Maximum number â€“ Minimum number of the data set

```python
np.max(data['Kms_Driven'])-np.min(data['Kms_Driven'])
```

##### b)	Variance: 
Square of difference between Mean with all data points

```python
data['Selling_Price'].var()
```

##### c) Standard Deviation: 
Square root of Variance

```python
sqrt(data['Selling_Price'].var())
```

##### d)	Skewness: 
This measure tells us whether the distribution of the data points is symmetric or skewed. Skewness gives us information about whether the data points are closed to center or spread across (tailed).
- If the right tailed skewness means, most of the data distribution is in left side and it is positive skewed
- If the left tailed skewness means, most of the data distribution is in right side and it is negative skewed  
- Skewness of the normal distribution is zero.


```python
lst = [10,20,30,10,20,40,10,20,30,10,20,40,10,20,30,10,20,40,20]
df = pd.DataFrame(lst)
```

```python
df.plot(kind = 'density')
print('This distribution has skew', df.skew())
```

##### e)	Kurtosis: 
It gives us the information as how peak your distribution is?
- If it is high peak means all the data points are very close to each other.
- If it is flattened means all the data points are spread across 


```python
lst2 = [10,20,30,20,20,20,20,20,20,20,20,20,20,20,20]
df = pd.DataFrame(lst2)
```

```python
df.plot(kind = 'density')
print('This distribution has kurtosis', df.kurt())
```

### Visualization Methods in Descriptive Statistics:
#### Categorical Variable:

##### With Frequency Distribution Table

```python
data['Year'].value_counts()
```

##### Bar Chart: Simple representation of graphical manner which contains Number of occurrences of each data point

```python
data['Year'].value_counts().plot(kind='bar');
```

```python
sns.set(font_scale=1.4)
data['Year'].value_counts().plot(kind='bar', figsize=(12, 6), rot=0)
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of Cars", labelpad=14)
plt.title("Count of Cars sold by Year", y=1.02);
```

##### Pie Chart: A pie chart is a type of graph that represents the data in the circular graph. The slices of pie shows the occurances of each data point

```python
data['Year'].value_counts().plot(kind='pie')
```

#### Numerical Variables:
##### Histogram: It contains the number of occurrences of range of data points. It creates range/buckets. It is used to know the distribution of dataset.

```python
plt.hist(data['Selling_Price'])
```

##### Box Plot: The main purpose is to find the outliers of data.

-	Median (Q2/50th Percentile): the middle value of the dataset.
-	First quartile (Q1/25th Percentile): the middle number between the smallest number (not the â€œminimumâ€) and the median of the dataset.
-	Third quartile (Q3/75th Percentile): the middle value between the median and the highest value (not the â€œmaximumâ€) of the dataset.
-	Interquartile range (IQR): 25th to the 75th percentile.
-	Whiskers (shown in blue):
-	â€œmaximumâ€: Q3 + 1.5*IQR
-	â€œminimumâ€: Q1 -1.5*IQR
-	Outliers: (shown as green circles)
-	We will get the outliers for a single feature(column) not entire row.

```python
plt.boxplot(data['Selling_Price'])
```

##### Scatter Plot: To represent two numerical variables on the same graph use scatter plot

```python
plt.scatter(data.Present_Price,data.Selling_Price)
```
